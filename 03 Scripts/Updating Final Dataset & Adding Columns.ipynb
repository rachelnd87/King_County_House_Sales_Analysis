### 01.Importing Libraries & Data

```python
#importing libraries
import pandas as pd
import numpy as np
import os
```

```python
#defining path to project folder
path=r'C:\Users\rache\Documents\House Sales Analysis'
```

```python
#importing house_sales_data_upd.csv
df=pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'house_sales_upd2.pkl'))
```

### 02.Creating House Age Column

```python
#creating new column for house_age using yr_built column as part of formula
df=df.assign(house_age=2022-df.yr_built)
df
```

### 03.Adding City & State Columns Using Zipcode

```python
import uszipcode #importing new library
from uszipcode import SearchEngine, SimpleZipcode, ComprehensiveZipcode #importing functions from new library
search = SearchEngine()

#defining function to find city using existing zipcode column
def zco(x):
    city = search.by_zipcode(x).major_city
    return city if city else 'None'

df['city'] = df['zipcode'].apply(zco)
```

```python
#redefining function to find state using existing zipcode column
def zco(x):
    state = search.by_zipcode(x).state
    return state if state else 'None'

df['state'] = df['zipcode'].apply(zco)
df
```

```python
#rearraging columns
df=df.reindex(columns=['Index', 'id', 'date', 'price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'yr_built','house_age','city','state','zipcode','lat', 'long', 
                       'Home Size Category','waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_renovated', 'sqft_living15', 'sqft_lot15'])
```

```python
#exporting data to csv
df.to_csv(os.path.join(path, '02 Data', 'Prepared Data', 'house_sales_final.csv'))
```
